```python
#looking up the number of people who have diabetes and insulin
filt=df['medical_specialty']== 'InternalMedicine'
df.loc[filt]['insulin'].str.contains('Up').count()
```

```python
# Looking up people that saw Internal Medicine on diabtes Meds and thier insulin was UP
df[(df['medical_specialty']=='InternalMedicine') & (df['insulin']=='Up')& (df['diabetesMed']=='Yes')].count()
```

```python
# Looking up people that saw Internal Medicine on diabtes Meds and thier insulin was Steady
df[(df['medical_specialty']=='InternalMedicine') & (df['insulin']=='Steady')& (df['diabetesMed']=='Yes')].count()
```

```python
# Looking up people that saw Internal Medicine, on diabtes Meds and thier insulin was Down
df[(df['medical_specialty']=='InternalMedicine') & (df['insulin']=='Down')& (df['diabetesMed']=='Yes')].count()
```
